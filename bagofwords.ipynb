```
import re
import matplotlib.pyplot as plt
mytext = "India is a country in South Asia. It is the seventh-largest country by area, the second-most populous country, and the most populous democracy in the world. Bounded by the Indian Ocean on the south, the Arabian Sea on the southwest, and the Bay of Bengal on the southeast, it shares land borders with Pakistan to the west;[e] China, Nepal, and Bhutan to the north; and Bangladesh and Myanmar to the east. In the Indian Ocean, India is in the vicinity of SriLanka and the Maldives; its Andaman and Nicobar Islands share a maritime border with Thailand and Indonesia"
len(mytext)
mytext=re.split("\\W+",mytext)
str2=[]
X=[]
Y=[]
for i in mytext:
  if i not in str2:
    str2.append(i)
    for i in range(0,len(str2)):
      print(str2[i]),mytext.count(str2[i])
      X.append(str2[i])
      Y.append(mytext.count(str2[i]))
```

```
fig=plt.figure(figsize=(20,10))
plt.bar(X,Y)
plt.xticks(rotation="vertical")
plt.show()
```
