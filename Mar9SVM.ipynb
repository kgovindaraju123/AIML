```
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC, LinearSVR
from sklearn import datasets
from sklearn import svm
from matplotlib import pyplot as plt
```

```
Iris =datasets.load_iris()
X=Iris.data[:100,:2]
y=Iris.target[:100]
#print(datasets)
```

```
#standardize features
scaler=StandardScaler()
X_std=scaler.fit_transform(X)
```

```
# Create and Train support vector classifier
svc = LinearSVC(C=1.0) #C is regularization parameter
#Train Model
model=svc.fit(X_std,y)
```

```
#Boundary
#plot data points and color using their class
color=['black' if C==0 else 'lightgrey' for C in y]
plt.scatter(X_std[:,0],X_std[:,1],c=color)

#Create the hyper plane
w=svc.coef_[0]
a=-w[0]/w[1]
xx=np.linspace(-2.5,2.5)
yy=a*xx -(svc.intercept_[0])/w[1]

#plot the hyperplane
plt.plot(xx,yy)
#plt.axis("off")
plt.show()
```

```
#using svm linear RBF Radial Basis Function
iris=datasets.load_iris()
X=iris.data[:,:2]
y=iris.target
#y #using three columns of data 0,1,2
```

```
#kernel,C and gamma are hyper parameters for SVM algorthm
svc=svm.SVC(kernel='linear',C=10000,gamma=1).fit(X,y)
```

```
#Create mesh plot for SVM
x_min, x_max=X[:,0].min()-1,X[:,0].max()+1
y_min, y_max=X[:,1].min()-1,X[:,1].max()+1
h=(x_max/x_min)/100
xx,yy =np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
```

```
plt.subplot(1,1,1)
Z=svc.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.contourf(xx,yy,Z,cmap=plt.cm.Paired,alpha=0.8)
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.xlim(xx.min(),xx.max())
plt.title('SVC with linear kernel')
plt.show()
```

```
svc=svm.SVC(kernel='rbf',C=10,gamma=4).fit(X,y)
#Create mesh plot for SVM
x_min, x_max=X[:,0].min()-1,X[:,0].max()+1
y_min, y_max=X[:,1].min()-1,X[:,1].max()+1
h=(x_max/x_min)/100
xx,yy =np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
plt.subplot(1,1,1)
Z=svc.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.contourf(xx,yy,Z,cmap=plt.cm.Paired,alpha=0.8)
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.xlim(xx.min(),xx.max())
plt.title('SVC with rbf kernel')
plt.show()
```
